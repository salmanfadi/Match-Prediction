```python
import requests
from bs4 import BeautifulSoup


url="https://www.statbunker.com/alltimestats/AllTimeLeagueTable?comp_code=EPL"
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')  # specify 'html.parser' as the parser
print(soup.prettify())
```

```python
soup.find('table')
```

```python

titles=soup.find_all('th')
titles
```

```python

table_titles=[info.text.strip() for info in titles[0:12]]
table_titles
```

```python
import pandas as pd
df=pd.DataFrame(columns=table_titles)
df
```

```python
rows=soup.find_all('tr')
rows
```

```python
for row in rows[1:]:
    row_data=row.find_all('td')
    row_details=[row1.text.strip() for row1 in row_data[0:12]]
   
    length=len(df)
    df.loc[length]=row_details
    
```
